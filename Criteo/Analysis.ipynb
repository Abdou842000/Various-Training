```python
import pandas as pd

# Convert Timestamp to datetime format for easier handling
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')

# Sort by user and timestamp to ensure chronological order
df = df.sort_values(by=['User ID', 'Timestamp'])

# Define a session window (e.g., 30 minutes) for grouping user activity
df['Session'] = df.groupby(['User ID'])['Timestamp'].diff().gt(pd.Timedelta(minutes=30)).cumsum()

# Filter out 'Sales' and 'Basket' events since they are not needed
df = df[~df['Event type'].isin(['Sales', 'Basket'])]

# Identify transitions from Listing to Product for the same user and session
df['Next Event'] = df.groupby(['User ID', 'Session'])['Event type'].shift(-1)
df['Next Product ID'] = df.groupby(['User ID', 'Session'])['Product ID'].shift(-1)

# Identify and remove Listing events that are followed by a Product event within the same session and for the same product
df_cleaned = df[~((df['Event type'] == 'Listing') & (df['Next Event'] == 'Product') & (df['Product ID'] == df['Next Product ID']))]

# Drop the auxiliary columns
df_cleaned = df_cleaned.drop(columns=['Next Event', 'Next Product ID'])

# Output the cleaned dataset
df_cleaned.info()
```

```python
# Filter the dataset for the period between 01/04/2016 and 31/03/2017
filtered_data_period = ecom[(ecom['Timestamp'] >= '2016-07-01') & (ecom['Timestamp'] <= '2016-07-31')]

# Separate visits (Product and Listing) and sales (Sales)
visits_data_period = filtered_data_period[filtered_data_period['Event type'].isin(['Product', 'Listing'])]
sales_data_period = filtered_data_period[filtered_data_period['Event type'] == 'Sales']

# Group visits and sales by date
visits_data_period['Date'] = visits_data_period['Timestamp'].dt.date
sales_data_period['Date'] = sales_data_period['Timestamp'].dt.date

# Count the number of visits per day
visits_per_day = visits_data_period.groupby('Date').size()

# Calculate the conversion rate per day (sales / visits)
sales_per_day = sales_data_period.groupby('Date').size()


# Plot the evolution of visits and conversion rate using two Y axes for the period
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot visits on the first Y axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Visits (Count)', color='tab:blue')
ax1.plot(visits_per_day.index, visits_per_day, color='tab:blue', label='Visits (Count)')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second Y axis for the conversion rate
ax2 = ax1.twinx()
ax2.set_ylabel('Conversion Rate', color='tab:red')
ax2.plot(sales_per_day.index, sales_per_day, color='tab:red', label='Sales')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title('Evolution of Product & Listing Events vs. Sales  (01/07/2016 - 30/07/2016)')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = ecom
# Convert Unix timestamp to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')

start_date = '2016-04-01'
end_date = '2017-04-01'
df = df[(df['Timestamp'] >= start_date) & (df['Timestamp'] < end_date)]

# Filter sales events and calculate revenue
sales_df = df[df['Event type'] == 'Sales']
sales_df['Revenue'] = sales_df['Product Price'] * sales_df['Product quantity']

# Group by month to calculate monthly revenue
monthly_revenue = sales_df.groupby(sales_df['Timestamp'].dt.to_period('M'))['Revenue'].sum()

# Plot revenue trend over the year
monthly_revenue.plot(kind='bar', figsize=(20,10), title='Monthly Revenue Over the Year')
plt.ylabel('Revenue ()')
plt.show()
```

```python
# Group by day to calculate the total revenue per day
daily_revenue = sales_df.groupby(sales_df['Timestamp'].dt.to_period('D'))['Revenue'].sum()

# Plot daily revenue trend with bars
ax = daily_revenue.plot(kind='bar', figsize=(15,7), title='Daily Revenue Over the Year')
plt.ylabel('Revenue ()')
plt.xlabel('Date')

# Create a list of the first and 15th day of each month for the x-ticks
first_and_15th_days = daily_revenue.index[daily_revenue.index.day.isin([1, 15])]

# Set custom x-ticks at the first and 15th day of each month
ax.set_xticks([daily_revenue.index.get_loc(day) for day in first_and_15th_days])  # Positioning the ticks
ax.set_xticklabels(first_and_15th_days, rotation=45)  # Labeling the ticks

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by product category and date (day)
sales_df['Date'] = sales_df['Timestamp'].dt.to_period('D')
category_groups = sales_df.groupby(['Product category', 'Date'])['Revenue'].sum().unstack(fill_value=0)

# Get the unique product categories
categories = sales_df['Product category'].unique()

# Create subplots with shared x-axis
fig, axes = plt.subplots(nrows=len(categories), ncols=1, figsize=(15, 15), sharex=True)

# Plot each product category separately but with shared x-axis
for i, category in enumerate(categories):
    category_groups.loc[category].plot(kind='bar', ax=axes[i], title=f'Revenue for {category}')
    axes[i].set_ylabel('Revenue ()')
    axes[i].set_xlabel('Date')
    
    # Display only a subset of x-axis labels to prevent overcrowding
    if i == len(categories) - 1:  # Show labels only on the last subplot
        axes[i].set_xticks(range(0, len(category_groups.columns), 30))  # Show labels every 30 days
        axes[i].set_xticklabels(category_groups.columns[::30], rotation=90)  # Rotate x-axis labels
    else:
        axes[i].set_xticks([])  # Hide x-axis labels for all but the last subplot

plt.tight_layout()
plt.show()
```

### Notes
- In Stressed periods where users prepare for a special purchase (black friday of christmas), they tend to use mode confortable/reliable devices such as tabets/ipads and desktops.
- From November to February, we notice a small increase in the usage of ipads and tablets, and significant decrease in the activity of mobile web browsing.
- The introduction of the android app impacted the overall contribution of device types to the traffic. Iphones and Ipads should not be impacted by this.
- We can notice that in the first weeks of introducing the android app, the desktop traffic contribution dropped, suggesting that users now try the android app as a new ergonomic and easily accessible way of browsing and shopping.
- The few first weeks act as a proof that mobile web browsing was not affected, probably because users going for mobile browsing don't bother with ergonomy, or at least when they decide to use mobile web browsing, they use it while shopping items not requiring this ergonomy.
- Another explanation in the drop of desktop usage then its increase during holidays, the launch was defenitely advertised, so the regular users knew it and decided to switch to the app for a more convenient browsing experience (loyalty, notification etc). The rise of desktop usage afterwards is probably due to occasional users, as they come just for special deals and holiday promotions. They may not even know about the app.
